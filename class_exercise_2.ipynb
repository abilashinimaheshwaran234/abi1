```python
class Member:
    def __init__(self, name, membershipID):
        self.name = name
        self.membershipID = membershipID

    def getBorrowLimit(self):
        return 0

    def display(self):
        print("Name:", self.name)
        print("Membership ID:", self.membershipID)


class RegularMember(Member):
    def getBorrowLimit(self):
        return 3

    def display(self):
        print("Member Type: Regular Member")
        super().display()
        print("Borrowing Limit:", self.getBorrowLimit())


class PremiumMember(Member):
    def getBorrowLimit(self):
        return 10

    def display(self):
        print("Member Type: Premium Member")
        super().display()
        print("Borrowing Limit:", self.getBorrowLimit())


def main():
    r1 = RegularMember("Abilashini", 101)
    p1 = PremiumMember("Deepashika", 202)

    r1.display()
    print()
    p1.display()


if __name__ == "__main__":
    main()
```
